```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hr-team.csv')
df
```

```python
df['2015-2021-fb'] = df['2015-fb'] + df['2016-fb'] + df['2017-fb'] + df['2018-fb'] + df['2019-fb'] + df['2021-fb']
```

```python
df
```

```python
df['2015-2021-hr'] = df['2015-hr'] + df['2016-hr'] + df['2017-hr'] + df['2018-hr'] + df['2019-hr'] + df['2021-hr']
```

```python
df['2015-21-hr%'] = df['2015-2021-hr']/df['2015-2021-fb']*100
```

```python
df['2022-hr%'] = df['2022-hr']/df['2022-fb']*100
```

```python
df['hr%-difference'] = df['2022-hr%'] - df['2015-21-hr%']
```

```python
df.sort_values('hr%-difference')
```
